```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv("D:\Review_db.csv\Review_db.csv")  # Fix the file path format (added 'r' for raw string)

# Apply Label Encoding to 'City' and 'Place'
encoder = LabelEncoder()
df['City'] = encoder.fit_transform(df['City'])
df['Place'] = encoder.fit_transform(df['Place'])

# Select relevant features
features = df[['City', 'Place']]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
df['Location_Segment'] = kmeans.fit_predict(features)

# Show first few results
print(df[['City', 'Place', 'Location_Segment']].head())
```

```python

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load dataset
df = pd.read_csv("D:\Review_db.csv\Review_db.csv")  # Ensure the file path is correct

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment
def get_sentiment(text):
    if pd.isna(text):  # Handle missing values
        return "Neutral"
    score = analyzer.polarity_scores(str(text))['compound']
    return "Positive" if score > 0 else "Negative" if score < 0 else "Neutral"

# Apply function to 'Raw_Review' column
df['Sentiment'] = df['Raw_Review'].astype(str).apply(get_sentiment)

# Show results
print(df[['Raw_Review', 'Sentiment']].head())
```

```python
# Apply K-Means clustering based on ratings
import pandas as pd

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 segments
df = pd.read_csv("D:\Review_db.csv\Review_db.csv")
df['Rating_Segment'] = kmeans.fit_predict(df[['Rating']])
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime format
df['Month'] = df['Date'].dt.month  # Extract the month
df['Year'] = df['Date'].dt.year  # Extract the year
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode categorical columns
encoder = LabelEncoder()
df['Sentiment'] = encoder.fit_transform(df['Sentiment'])

# Select relevant features for clustering
features = df[['City', 'Place', 'Rating', 'Sentiment', 'Month']]
features = pd.get_dummies(features)  # Convert categorical data

# Normalize data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
df['Customer_Segment'] = kmeans.fit_predict(features_scaled)
```
